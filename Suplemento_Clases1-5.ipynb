# Instrucciones de flujo

Las instrucciones de flujo permiten imponer condiciones lógicas de manera que, en caso de ser Verdaderas o Falsas (según determinemos) definirán qué líneas de código se ejecutan. Pensemos en la acción de comprar un computador, para el cual tenemos un presupuesto de solo 700 USD.

## Ejemplo:


```python
presupuesto=600.0

#Precios de diferentes computadores
precioPC1=500.0
precioPC2=450.0
precioPC3=700.1
```

```python
if(presupuesto>=precioPC1):
    print("Compré el PC1")
elif(presupuesto<precioPC1):
    print("No compre el PC1")
else:
    print("Esta opción no sucederá")
```

```python
if(presupuesto>=precioPC2):
    print("Compré el PC2")
elif(presupuesto<precioPC2):
    print("No compre el PC2")
else:
    print("Esta opción no sucederá")
```

```python
if(presupuesto>=precioPC3):
    print("Compré el PC2")
elif(presupuesto<precioPC3):
    print("No compre el PC2")
else:
    print("Esta opción no sucederá")
```

También podemos tener condiciones anidadas, por ejemplo, es posible que además de pensar en qué computador esté dentro de nuestro presupuesto, tengamos que pensar en que debemos pagar 110 extra por impuestos en cualquiera de los casos. Probemos ahora.

```python
if(presupuesto>=precioPC1):
    if(presupuesto-110>=precioPC1):
        print("Compré el PC1")
    elif(presupuesto-110<precioPC1):
        print("No compre el PC1 por impuestos")
elif(presupuesto<precioPC1):
        print("No compre el PC1 por el precio")
```

Claramente pudimos evalar desde el principio si el presupuesto menos los impuestos eran suficiente para pagar el PC1, sin embargo, anidando las condiciones podemos determinar de manera exacta por qué no lo pudimos comprar

# Listas y ciclos
Otras instrucciones de flujo son los ciclos. Sin profundizar en las listas por ahora, creemos una lista que contenga los precios y evaluemos. Repitamos el ejercicio, pero ahora evaluemos de manera iterativa las condiciones.

```python
lista=[precioPC1,precioPC2,precioPC3]
#La lista tiene 3 elementos, podemos imprimir el índice de
#cada elemento de la lista, de la siguiente manera
for i in range(len(lista)):
    print(i)
```

```python
#Por otro lado, podemos imprimir los elementos de la lista, como tal, así:
print("Accediendo por índice")
for i in range(len(lista)):
    print(lista[i])
    
#o
print("\n")
print("Accediendo directamente al elemento")
for valor in lista:
    print(valor)
```

Ahora, realicemos la operación que habiamos planteado de evaluar si podíamos comprar un PC, pero escribiendo una única vez la instrucción y evaluandola con un ciclo para cada PC en la lista

```python
for i in range(len(lista)):
    precioPCi=lista[i]
    if(presupuesto>=precioPCi):
        if(presupuesto-110>=precioPCi):
            print("Compré el PC%d"%(i+1))
        elif(presupuesto-110<precioPCi):
            print("No compre el PC%d por impuestos"%(i+1))
    elif(presupuesto<precioPCi):
        print("No compre el PC%d por el precio"%(i+1))
```

# While: break y continue

```python
i,j=0,0        #Empezaremos i y j como 0

while(i==0):   #Siempre que i sea igual a cero
    print(j)   #Imprimirá j
    j+=1       #j aumentará en una unidad cada iteración
    if(j==10):
        break  #Terminará el ciclo cuando j llegue a 10
```

```python
i=0
while(i<8):            #Mientras que i sea menor a 8
    i+=1               #Aumentará i en 1 cada iteración
    if(i==5): continue #Pasará a la siguiente iteración si se cumple.
    print(i)
```

```python
#Lo último es el comando pass, que sirve para no hacer nada durante una iteración
for i in 'Palabra':
    pass
print(i)
```
